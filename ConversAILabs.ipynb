```python
!pip install -U openai-whisper
import whisper
# Load the Whisper model
model = whisper.load_model("base")  # "small" is more accurate, slower
# Replace with uploaded file name
audio_file = list(uploaded.keys())[0]
# Transcribe audio
result = model.transcribe(audio_file)
# Extract transcript
transcript = result['text']
print("ðŸ“„ Full Transcript:\n")
print(transcript)
```

```python
from textblob import TextBlob
import time
start_time = time.time()
# Tokenize transcript into sentences
blob = TextBlob(transcript)
sentences = [str(sentence) for sentence in blob.sentences]
# Alternate speaker turns (simulated)
speaker1, speaker2 = [], []
current_speaker = 1
for sentence in sentences:
    if current_speaker == 1:
        speaker1.append(sentence)
        current_speaker = 2
    else:
        speaker2.append(sentence)
        current_speaker = 1
# Word counts
speaker1_words = sum(len(s.split()) for s in speaker1)
speaker2_words = sum(len(s.split()) for s in speaker2)
total_words = speaker1_words + speaker2_words
speaker1_ratio = round((speaker1_words / total_words) * 100, 2)
speaker2_ratio = round((speaker2_words / total_words) * 100, 2)
print(f"\n Talk-time ratio:\n  Speaker 1: {speaker1_ratio}%\n  Speaker 2: {speaker2_ratio}%")
num_questions = transcript.count("?")
print(f"\n Number of questions asked: {num_questions}")
# Simple heuristic for speaker detection using names
sentences = transcript.split('. ')  # basic split
monologues = []
current_speaker = None
current_text = ""
for sentence in sentences:
    if "Lauren" in sentence or "I can help" in sentence:
        speaker = 1
    elif "John" in sentence or "I have" in sentence:
        speaker = 2
    else:
        speaker = current_speaker  # continue same speaker
    if speaker == current_speaker:
        current_text += " " + sentence
    else:
        if current_text:
            monologues.append((current_speaker, current_text.strip()))
        current_text = sentence
        current_speaker = speaker
monologues.append((current_speaker, current_text.strip()))
# Calculate longest monologue
monologue_lengths = [len(text.split()) for _, text in monologues]
longest_monologue = max(monologue_lengths)
approx_seconds = round((longest_monologue / 15) * 5, 2)
print(f"\n Longest monologue duration (approx.): {approx_seconds} seconds")
polarity = blob.sentiment.polarity
if polarity > 0.1:
    sentiment_label = "Positive"
elif polarity < -0.1:
    sentiment_label = "Negative"
else:
    sentiment_label = "Neutral"
print(f"\n Call sentiment: {sentiment_label} (score: {polarity})")
if num_questions < 5:
    insight = "Ask more open-ended questions to better understand customer needs."
else:
    insight = "Good use of questions. Keep guiding the conversation naturally."
print(f"\n One actionable insight: {insight}")
sales_keywords = ["order", "promotion", "vehicle", "price", "shipping", "update", "offer", "credit card"]
# Score each speaker separately
speaker1_score = sum([1 for s in speaker1 if any(kw in s.lower() for kw in sales_keywords)])
speaker2_score = sum([1 for s in speaker2 if any(kw in s.lower() for kw in sales_keywords)])
likely_sales_rep = "Speaker 1" if speaker1_score > speaker2_score else "Speaker 2"
print(f"\n Likely Sales Rep: {likely_sales_rep}")
```
